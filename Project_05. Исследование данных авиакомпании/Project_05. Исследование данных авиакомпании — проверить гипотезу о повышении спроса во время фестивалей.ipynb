## Шаг 5. Проверка гипотезы

Проверим гипотезы о влиянии проводимых фестивалей на средний спрос на авиабилеты.

```python
query_last = pd.read_csv('/datasets/query_last.csv')
query_last.info()
```

```python
query_last.head()
```

Выведем два среза: строки без фестивалей и с фестивалями.

```python
query_last['festival_name'] = query_last['festival_name'].fillna('none')
no_festival = query_last.query('festival_name == "none"')
no_festival
```

```python
festival = query_last.query('festival_name != "none"')
festival
```

**Сформулируем гипотезы**

<br> H₀: Средний спрос на билеты во время фестивалей не отличается от среднего спроса на билеты в обычное время.
<br> H₁: Средний спрос на билеты во время фестивалей отличается от среднего спроса на билеты в обычное время.
<br> Пороговое значение alpha (критический уровень статистической значимости) зададим равным 5%.

```python
ticket_festival = festival['ticket_amount']
ticket_no_festival = no_festival['ticket_amount']
alpha = .05 
results = st.ttest_ind(
    ticket_festival, 
    ticket_no_festival)
print('p-значение:', results.pvalue)
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

***Вывод по шагу 5:***
<br> P-value равно 0.097 и у нас не получается отвергнуть нулевую гипотезу. Если бы выборки на самом деле были одинаковыми, то почти в 10% случаев различия между выборками получились случайно. Наличие фестивалей не влияет на количество купленных билетов.

<br> H₀: Количество рейсов в ТОП-10 не отличается от количества рейсов в остальных городах.
<br> H₁: Количество рейсов в ТОП-10 отличается от количества рейсов в остальных городах.
<br> Пороговое значение alpha (критический уровень статистической значимости) зададим равным 5%.

```python
alpha = 0.05 # критический уровень статистической значимости
             # если p-value окажется меньше него - отвергнем гипотезу

top_10 = city[city['city'].isin(top10_cities['city'])]
other_cities = city[~city['city'].isin(top10_cities['city'])]

alpha = .05 
results = st.ttest_ind(top_10['average_flights'], other_cities['average_flights'], equal_var=False)
print('p-значение:', results.pvalue)
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

***Вывод по шагу 5:***
<br> P-value равно 0.099 и здесь также не получилось отвергнуть нулевую гипотезу. Количество рейсов в ТОП-10 не отличается от количества рейсов в остальных городах.

```python
def average_flights_group(row):
    group = row['average_flights']
    if group <= 5:
        return 'менее 5 рейсов'
    if group >= 5 and group < 10:
        return '5-10 рейсов'
    return 'более 10'

city['average_flights'] = city.apply(average_flights_group, axis=1) 
```

```python
city_from_average_flights = pd.DataFrame()
city_from_average_flights['count_average_flights'] = city.groupby('average_flights')['city'].count()
city_from_average_flights.sort_values('count_average_flights', ascending = False)
```

Произвел разбивку по количеству рейсов в городах. Разбивку по квантилям и Топ3-5 не использовал, т.к. считаю будет не очень информативным. А так в среднем по нашей выборке совершается около 5 рейсов, от минимального - 1 рейс до максимального - 129 рейсов. До 25% - 2 рейса, 50% - 3 рейса, 75% - 5 рейсов.
